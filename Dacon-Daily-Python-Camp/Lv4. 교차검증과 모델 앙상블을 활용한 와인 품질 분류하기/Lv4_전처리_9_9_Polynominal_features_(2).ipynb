```python
# ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë§í¬ë¡œ ë°ì´í„°ë¥¼ ì½”ë©ì— ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd

# PolynomialFeatures ë¼ì´ë¸ŒëŸ¬ë¦¬ í˜¸ì¶œ
from sklearn.preprocessing import PolynomialFeatures

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

# ë²”ì£¼í˜• ë³€ìˆ˜ ì œê±°
train.drop('type', axis = 1,inplace = True)
test.drop('type', axis = 1,inplace = True)

# index ë³€ìˆ˜ ì œê±°

train.drop('index', axis = 1,inplace = True)
test.drop('index', axis = 1,inplace = True)
```

## ğŸ‘‹ ì‹¤ìŠµ
---
ì´ë²ˆ ì‹œê°„ì—ëŠ” ì™€ì¸ í’ˆì§ˆ ë¶„ë¥˜ ë°ì´í„°ì— Polynomial Featuresë¥¼ ì ìš©í•´ ë³´ê³  decision tree ëª¨ë¸ì„ ì´ìš©í•´ í’ˆì§ˆ ë¶„ë¥˜ë¥¼ í•´ë³´ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤. 

1. train ë°ì´í„° ë³€í™˜
2. ëª¨ë¸ í•™ìŠµ (Decision Tree)
3. test ë°ì´í„° ë³€í™˜
4. ì¶”ë¡ 
5. ì •ë‹µ íŒŒì¼ ìƒì„±

```python
# ë°ì´í„° í™•ì¸
train.head()
```

```python
 # train ë°ì´í„°ë¥¼ PolynomialFeatures ë¥¼ ì´ìš©í•˜ì—¬ ë³€í™˜
poly_features = PolynomialFeatures(degree=2)

df = train.drop('quality', axis=1)
df_poly = poly_features.fit_transform(df)
df_poly = pd.DataFrame(df_poly)
```

```python
# DecisionTreeClassifier ëª¨ë¸ì„ ë³€í™˜ëœ train ë°ì´í„°ë¡œ í•™ìŠµ

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(df_poly,train['quality'])
```

```python
# test ë°ì´í„° ë³€í™˜
poly_features = PolynomialFeatures(degree=2)
test_poly = poly_features.fit_transform(test)
test_poly = pd.DataFrame(test_poly)
```

```python
# ê²°ê´ê°’ ì¶”ë¡ 
pred = model.predict(test_poly)
```

```python
# ì •ë‹µ íŒŒì¼ ìƒì„±
submission = pd.read_csv('data/sample_submission.csv')
submission['quality'] = pred
submission.to_csv('poly.csv', index=False)
```

```python
 # train ë°ì´í„°ë¥¼ PolynomialFeatures ë¥¼ ì´ìš©í•˜ì—¬ ë³€í™˜
 
poly_features = PolynomialFeatures(degree=2) # ì°¨ì›ì€ 2ë¡œ ì„¤ì •

# ì™€ì¸ í’ˆì§ˆ ê¸°ì¤€ì¸ quality ë³€ìˆ˜ë¥¼ ì œì™¸í•œ ë‚˜ë¨¸ì§€ ë³€ìˆ˜ë¥¼ í¬í•¨í•œ ë°ì´í„° ë³€í™˜.
df = train.drop('quality',axis = 1)
df_poly = poly_features.fit_transform(df) # fit_transform ë©”ì†Œë“œë¥¼ í†µí•´ ë°ì´í„° ë³€í™˜
df_poly = pd.DataFrame(df_poly) # PolynomialFeaturesë¡œ ë³€í™˜ ëœ ë°ì´í„°ë¥¼ ë°ì´í„° í”„ë ˆì„ í˜•íƒœë¡œ ë³€í™˜
```
