```python
# ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë§í¬ë¡œ ë°ì´í„°ë¥¼ ì½”ë©ì— ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
from lightgbm import LGBMClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

## ğŸ‘‹ ì‹¤ìŠµ
---


```python
# ë°ì´í„° í™•ì¸
train.head()
```

```python
# ì›í•« ì¸ì½”ë”© (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# ëª¨ë¸ ì •ì˜
model = LGBMClassifier()
```

```python
# ëª¨ë¸ í•™ìŠµ

# X ëŠ” trainì—ì„œ quality ë¥¼ ì œì™¸í•œ ëª¨ë“  ë³€ìˆ˜
X = train_one.drop('quality', axis=1)

# y ëŠ” trainì˜ quality ë³€ìˆ˜
y = train_one['quality']

# fit ë©”ì†Œë“œë¥¼ ì´ìš©í•´ ëª¨ë¸ í•™ìŠµ
model.fit(X, y)
```

```python
# predict ë©”ì†Œë“œì™€ test_one ë°ì´í„°ë¥¼ ì´ìš©í•´ í’ˆì§ˆ ì˜ˆì¸¡
pred = model.predict(test_one)
```

```python
# sample_submission.csv íŒŒì¼ì„ ë¶ˆëŸ¬ì™€ ì˜ˆì¸¡ëœ ê°’ìœ¼ë¡œ ì±„ì›Œ ì£¼ê¸°
submission = pd.read_csv('data/sample_submission.csv')
submission['quality'] = pred
submission.head()
```

```python
# ì •ë‹µíŒŒì¼ ë‚´ë³´ë‚´ê¸°
submission.to_csv('lgbm_pred.csv', index = False)
```
